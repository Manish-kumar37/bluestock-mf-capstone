```python
from pathlib import Path

raw_path = Path("../data/raw")

files = []

for file in raw_path.glob("*"):
    files.append(
        {
            "file_name": file.name,
            "size_kb": round(file.stat().st_size / 1024, 2)
        }
    )

pd.DataFrame(files)
```
